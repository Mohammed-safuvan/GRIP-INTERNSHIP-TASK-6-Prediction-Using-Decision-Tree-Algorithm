# Importing the libaraies


   
   * The Python Seaborn library lets you visualize data using pair plots that produce a matrix of relationships between each        variable in the dataset.
   
   * Python Matplotlib is a library which basically serves the purpose of Data Visualization
   
   * Import the Decision Tree Libraries and its Functions.
   
   * Import Classification report,Accuracy Score,Confusion matrix to Analyse the Performance of the Model From Sklearn Metrics


```python
import pandas as pd
import seaborn as sns
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import make_scorer, f1_score, recall_score, precision_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

# Load the Iris Dataset
  * Load the Iris Dataset at any Variable
  * Loading this Dataset and Store it in a Dataframe

```python
data=load_iris()
ds=pd.DataFrame(data.data,columns=data.feature_names)
data.feature_names
```

# Fixing Target variable
  * The head () in Python returns the series with the first few rows (by default 5)

```python
ds['target']=data['target']
ds.head()
```

# Spliting the Dataset

   * train_test_split() method is available in the sklearnlibrary to split the data into train test sets.
   
   * Training dataset use to teach a machine learning application to recognize patterns.
   
   * Testing dataset is used to  evaluate the model's accuracy.

```python
x=data.target
y=data.target
x_train,x_test,y_train,y_test=train_test_split(ds[data.feature_names],ds['target'])
print("xtrain:",len(x_train))
print("xtest:",len(x_test))
```

# Using a Decision tree classifier to build the model

   * Use Decision tree classifier(criterion='gini')
   
   * The Purpose of gini is  decrease the level of entropy from the root nodes to the leaf nodes of the decision tree.
   
   

```python
# DT= DecisionTreeClassifier()
DT=DecisionTreeClassifier(criterion='gini')

# fit model
model=DT.fit(x_train,y_train)
y_preds=model.predict(x_test)
print(y_preds)
```

#  Plot the Decision Tree and Confusion matrix
   *  A confusion matrix is a table that is often used to describe the performance of a classification model  on a set of test       data for which the true values are known.
  
    

```python
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(7,5),dpi=500)
tree.plot_tree(DT,fontsize=5,feature_names=data.feature_names,class_names=data.target_names)

labels=[0,1,2]
cmx=confusion_matrix(y_test,y_preds, labels)
print("\n Confusion Matrix: \n",cmx)
print("\n")
print(classification_report(y_test,y_preds))
```
